```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
population_size=100000  #We are taking population as 1,00,000
sample_size=50          #To predict avg salary of population i am taking sample size as 50
num_samples=1000        #no. of samples is 1000
```

```python
np.random.seed(42)      # generating random numbers
```

```python
#assuming population follows lognormal distribution and generating 1,00,000 peoples salary
population_salaries=np.random.lognormal(mean=4.5,sigma=0.8,size=population_size)
```

```python
#selecting random numbers of size 50 from population_salaries and taking 1000 set of such sample
sample_means=[]
sample_std=[]
for i in range(num_samples):
    sample_salaries=np.random.choice(population_salaries,size=sample_size)
    sample_means.append(np.mean(sample_salaries))
    sample_std.append(np.std(sample_salaries))
```

```python
avg_sample_means=np.mean(sample_means)                      #mean of all 1000 set
act_sample_std=np.std(sample_means)/np.sqrt(sample_size)    #std dev of 1000 set
lower_limit=avg_sample_means-1.96*act_sample_std            #calculating lower limit by considering 95% confidence interval
lower_limit
```

```python
higher_limit=avg_sample_means+1.96*act_sample_std           #calculating lower limit by considering 95% confidence interval
higher_limit
```

```python
np.mean(population_salaries)     #population mean salary that here we know but not in real life
```

```python
avg_sample_means                 # sample means that we are getting
```

```python
print('Avg salaries based on population data of',population_size, 'is',np.mean(population_salaries))
print('Avg salaries that we predict based on sample of size',sample_size,'is',avg_sample_means)
print('Salary range should be between',lower_limit,'and ',higher_limit)
```
